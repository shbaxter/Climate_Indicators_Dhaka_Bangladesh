```python
baseline_df = df.loc[:, '2010':'2015']
baseline_avg = baseline_df.mean(axis=1)

baseline_avg_np = baseline_avg.to_numpy().reshape(-1, 1)

diff_df = df.loc[:, '2016':] - baseline_avg_np

plt.figure(figsize=(12, 6))

for year_col in diff_df.columns:
    plt.plot(df['Week'], diff_df[year_col], marker='o', label=year_col)

plt.xlabel('Week')
plt.ylabel('Difference from Baseline Average')
plt.title('Change in Values Compared to Baseline Average')  
plt.xticks(rotation=45) 
plt.grid(True)  
plt.legend(loc='upper left')  
plt.tight_layout() 
plt.show() 
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import ruptures as rpt

# Prepare data
data = df.iloc[:, 1:].values.flatten()  # Convert to 1D array
years = np.repeat(np.arange(2010, 2024), len(df))  # Generate years

# Perform change point detection using PELT algorithm
model = "l2"  # "l1" for absolute value, "l2" for squared error
algo = rpt.Pelt(model=model).fit(data)
result = algo.predict(pen=10)  # Adjust the penalty parameter for desired sensitivity

# Plot
plt.figure(figsize=(12, 6))
plt.plot(years, data, color='blue')
for r in result:
    if r < len(years):  # Ensure index is within bounds
        plt.plot(years[r], data[r], marker='o', markersize=8, color='red')
plt.title('Change Point Detection for Dengue Fever Cases (2010-2023)')
plt.xlabel('Year')
plt.ylabel('Dengue Fever Cases')
plt.grid(True)
plt.show()
```
